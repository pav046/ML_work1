```python
def ascii_difference(s):
    if not s:
        return 0

    ascii_values = [ord(char) for char in s]

    max_ascii = max(ascii_values)
    min_ascii = min(ascii_values)

    difference = max_ascii - min_ascii
    return difference


input_string = "Hello, World!"
result = ascii_difference(input_string)
print(result)
```

```python
def s(s):
    upper_string = s.upper()

    sorted_string = ''.join(sorted(upper_string))

    return sorted_string

input_string = "Hello, World!"
result = s(input_string)
print(result)
```

```python
import numpy as np

def my_2d_array(arr):
    n = 4
    shape = (len(arr) - n + 1, n)
    strides = (arr.strides[0], arr.strides[0])
    return np.lib.stride_tricks.as_strided(arr, shape=shape, strides=strides)


A = np.array([1, 2, 3, 4, 5])
B = my_2d_array(A)
print(B)
```

```python
import numpy as np

def h(arr, num_bits=8):
    t = np.array([list(np.binary_repr(x, width=num_bits)) for x in arr], dtype=int)
    return t

A = np.array([0, 1, 2])
B = h(A)
print(B)
```

```python
import pandas as pd




df['Ram'] = df['Ram'].astype(str)
df['Ram'] = df['Ram'].str.replace('GB', '').astype(int)

df['Weight'] = df['Weight'].str.replace('kg', '')


df[['Width', 'Height']] = df['ScreenResolution'].str.extract(r'(\d+)x(\d+)$')
df['Width'] = df['Width'].astype(int)
df['Height'] = df['Height'].astype(int)


df[['CpuType', 'CpuFreq']] = df['Cpu'].str.extract(r'(.+) (\d+\.\d+)GHz')
df['CpuFreq'] = df['CpuFreq'].astype(float)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(16, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    sns.boxplot(df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()


    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    if not outliers.empty:
        print(f'Выбросы для {col}:')
        print(outliers)
```
